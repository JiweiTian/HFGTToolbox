```matlab
function myLFES = setupServices(myLFES, S)
disp(['I am entering ' 'setupServices.m'])
myLFES.services=getResourceAttributes(S.LFES.Service);
[myLFES.services.('names')] = myLFES.services.('name');
myLFES.services = rmfield(myLFES.services,'name');
myLFES.services.servicePlaces = {};

for k1=1:length(S.LFES.Service)
    %% Check number of services
    if length(S.LFES.Service) == 1
        % Number of service = 1
        myLFES.services.servicePlaces{k1,1}=struct();
        if ismember('ServicePlace',fieldnames(S.LFES.Service))
            numServicePlaces = length(S.LFES.Service.ServicePlace);
            if numServicePlaces == 1
                myLFES.services.servicePlaces{k1,1}.names{1,1} = S.LFES.Service.ServicePlace.Attributes.name;
                if ismember('initialCondition',fieldnames(S.LFES.Service.ServicePlace.Attributes))
                    myLFES.services.servicePlaces{k1,1}.initialConditions{1,1} = S.LFES.Service.ServicePlace.Attributes.initialCondition;
                    myLFES.services.servicePlaces{k1,1}.finalConditions{1,1} = S.LFES.Service.ServicePlace.Attributes.finalCondition;
                end
            else
                myLFES.services.servicePlaces{k1,1} = struct();
                for k2 = 1:numServicePlaces
                    myLFES.services.servicePlaces{k1,1}.names{k2,1} = S.LFES.Service.ServicePlace{k2}.Attributes.name;
                    if ismember('initialCondition',fieldnames(S.LFES.Service.ServicePlace{k2}.Attributes))
                        myLFES.services.servicePlaces{k1,1}.initialConditions{k2,1} = S.LFES.Service.ServicePlace{k2}.Attributes.initialCondition;
                        myLFES.services.servicePlaces{k1,1}.finalConditions{k2,1} = S.LFES.Service.ServicePlace{k2}.Attributes.finalCondition;
                    end
                end
            end
        end
        myLFES.services.serviceTransitions{k1,1}=struct();
        if ismember('ServiceTransition',fieldnames(S.LFES.Service))
            numServiceTransitions = length(S.LFES.Service.ServiceTransition);
            if numServiceTransitions == 1
                myLFES.services.serviceTransitions{k1,1}=getResourceAttributes(S.LFES.Service.ServiceTransition);
            else
                for k2 = 1:numServiceTransitions
                    myLFES.services.serviceTransitions{k1,1}.methodLinkName{k2,1}=S.LFES.Service.ServiceTransition{k2}.Attributes.methodLinkName;
                    myLFES.services.serviceTransitions{k1,1}.methodLinkRef{k2,1}=S.LFES.Service.ServiceTransition{k2}.Attributes.methodLinkRef;
                    myLFES.services.serviceTransitions{k1,1}.name{k2,1}=S.LFES.Service.ServiceTransition{k2}.Attributes.name;
                    myLFES.services.serviceTransitions{k1,1}.postset{k2,1}=S.LFES.Service.ServiceTransition{k2}.Attributes.postset;
                    myLFES.services.serviceTransitions{k1,1}.preset{k2,1}=S.LFES.Service.ServiceTransition{k2}.Attributes.preset;
                end
            end
        end
    else
        % Number of services > 1
        %% Setup Service Places
        myLFES.services.servicePlaces{k1,1}=struct();
        if ismember('ServicePlace',fieldnames(S.LFES.Service{k1}))
            numServicePlaces = length(S.LFES.Service{k1}.ServicePlace);
            if numServicePlaces == 1
                myLFES.services.servicePlaces{k1,1}.names{1,1} = S.LFES.Service{k1}.ServicePlace.Attributes.name;
                if ismember('initialCondition',fieldnames(S.LFES.Service{k1}.ServicePlace.Attributes))
                    myLFES.services.servicePlaces{k1,1}.initialConditions{1,1} = S.LFES.Service{k1}.ServicePlace.Attributes.initialCondition;
                    myLFES.services.servicePlaces{k1,1}.finalConditions{1,1} = S.LFES.Service{k1}.ServicePlace.Attributes.finalCondition;
                end
            else
                myLFES.services.servicePlaces{k1,1} = struct();
                for k2 = 1:numServicePlaces
                    myLFES.services.servicePlaces{k1,1}.names{k2,1} = S.LFES.Service{k1}.ServicePlace{k2}.Attributes.name;
                    if ismember('initialCondition',fieldnames(S.LFES.Service{k1}.ServicePlace{k2}.Attributes))
                        myLFES.services.servicePlaces{k1,1}.initialConditions{k2,1} = S.LFES.Service{k1}.ServicePlace{k2}.Attributes.initialCondition;
                        myLFES.services.servicePlaces{k1,1}.finalConditions{k2,1} = S.LFES.Service{k1}.ServicePlace{k2}.Attributes.finalCondition;
                    end
                end
            end
        end
        %% Setup Service Transitions
        myLFES.services.serviceTransitions{k1,1}=struct();
        if ismember('ServiceTransition',fieldnames(S.LFES.Service{k1}))
            numServiceTransitions = length(S.LFES.Service{k1}.ServiceTransition);
            if numServiceTransitions == 1
                myLFES.services.serviceTransitions{k1,1}=getResourceAttributes(S.LFES.Service{k1}.ServiceTransition);
            else
                for k2 = 1:numServiceTransitions
                    myLFES.services.serviceTransitions{k1,1}.methodLinkName{k2,1}=S.LFES.Service{k1}.ServiceTransition{k2}.Attributes.methodLinkName;
                    myLFES.services.serviceTransitions{k1,1}.methodLinkRef{k2,1}=S.LFES.Service{k1}.ServiceTransition{k2}.Attributes.methodLinkRef;
                    myLFES.services.serviceTransitions{k1,1}.name{k2,1}=S.LFES.Service{k1}.ServiceTransition{k2}.Attributes.name;
                    myLFES.services.serviceTransitions{k1,1}.postset{k2,1}=S.LFES.Service{k1}.ServiceTransition{k2}.Attributes.postset;
                    myLFES.services.serviceTransitions{k1,1}.preset{k2,1}=S.LFES.Service{k1}.ServiceTransition{k2}.Attributes.preset;
                end
            end
        end
    end
end
disp(['I am leaving  ' 'setupServices.m'])
```
